```python
import pandas as pd
import numpy as np

# Cargar los datos
df = pd.read_csv('retail_sales_dataset.csv')

# Convertir la columna de fecha a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])
df.head(10)
```

```python
## 1. Transformación de Datos

# Crear nuevas columnas útiles para el análisis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek  # 0=Lunes, 6=Domingo
df['Quarter'] = df['Date'].dt.quarter
df['WeekdayName'] = df['Date'].dt.day_name()
```

```python
# La columna 'Total Amount' ya existe, pero verificamos que sea correcta
df['Calculated_Total'] = df['Quantity'] * df['Price per Unit']
# Verificamos si hay discrepancias
df['Total_Check'] = np.where(df['Total Amount'] == df['Calculated_Total'], 'OK', 'Mismatch')
```

```python
df.info()
```

```python
# Clasificar las ventas en categorías (Alta, Media, Baja) basado en el monto total
percentiles = df['Total Amount'].quantile([0.33, 0.66]).values
df['Sales_Category'] = pd.cut(df['Total Amount'],
                             bins=[0, percentiles[0], percentiles[1], float('inf')],
                             labels=['Low', 'Medium', 'High'])

# Crear categorías de edad
bins = [0, 25, 40, 55, 100]
labels = ['<25', '25-40', '40-55', '55+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
## 2. Agrupación y Agregación

# Agrupación por categoría de producto y agregación de estadísticas
product_category_stats = df.groupby('Product Category').agg({
    'Quantity': ['sum', 'mean', 'count', 'max', 'min', 'std'],
    'Total Amount': ['sum', 'mean', 'max', 'min'],
    'Price per Unit': ['mean', 'max', 'min']
}).reset_index()

# Agrupación por mes y categoría de producto
monthly_product_stats = df.groupby(['Year', 'Month', 'Product Category']).agg({
    'Total Amount': 'sum',
    'Quantity': 'sum',
    'Transaction ID': 'count'
}).reset_index()
monthly_product_stats.rename(columns={'Transaction ID': 'Transaction_Count'}, inplace=True)

# Agrupación por género y categoría de edad
gender_age_stats = df.groupby(['Gender', 'Age_Group']).agg({
    'Total Amount': ['sum', 'mean'],
    'Quantity': 'mean',
    'Transaction ID': 'count'
}).reset_index()
gender_age_stats.rename(columns={'Transaction ID': 'Transaction_Count'}, inplace=True)

# Agrupación por día de la semana
weekday_stats = df.groupby(['WeekdayName', 'DayOfWeek']).agg({
    'Total Amount': 'sum',
    'Quantity': 'sum',
    'Transaction ID': 'count'
}).reset_index().sort_values('DayOfWeek')
weekday_stats.rename(columns={'Transaction ID': 'Transaction_Count'}, inplace=True)

## 3. Análisis Personalizado con apply

# Función para calcular la desviación respecto a la media del grupo
def deviation_from_group_mean(row, group_col, value_col):
    group_mean = df.groupby(group_col)[value_col].mean()[row[group_col]]
    return (row[value_col] - group_mean) / group_mean * 100  # Porcentaje de desviación

# Aplicar la función para desviación de precios por categoría de producto
df['Price_Deviation_Pct'] = df.apply(
    lambda x: deviation_from_group_mean(x, 'Product Category', 'Price per Unit'),
    axis=1)

# Función para clasificar clientes según frecuencia de compra (asumiendo que Customer ID es único por cliente)
customer_frequency = df['Customer ID'].value_counts().reset_index()
customer_frequency.columns = ['Customer ID', 'Purchase_Count']
df = df.merge(customer_frequency, on='Customer ID', how='left')

# Clasificar clientes en categorías según frecuencia de compra
df['Customer_Type'] = pd.cut(df['Purchase_Count'],
                            bins=[0, 1, 3, float('inf')],
                            labels=['One-Time', 'Occasional', 'Frequent'])

# 4. Análisis de Valor por Cliente

# Calcular valor total por cliente
customer_value = df.groupby('Customer ID').agg({
    'Total Amount': 'sum',
    'Quantity': 'sum',
    'Transaction ID': 'count'
}).reset_index()
customer_value.columns = ['Customer ID', 'Total_Spent', 'Total_Items', 'Transaction_Count']

# Clasificar clientes por valor (RFM simplificado)
customer_value['Value_Category'] = pd.qcut(customer_value['Total_Spent'],
                                         q=4,
                                         labels=['Low', 'Medium', 'High', 'VIP'])

#5. Análisis de Precios

# Estadísticas de precios por categoría de producto
price_analysis = df.groupby('Product Category').agg({
    'Price per Unit': ['mean', 'median', 'std', 'min', 'max'],
    'Total Amount': 'sum'
}).reset_index()

# Relación entre precio y cantidad vendida
price_quantity_corr = df.groupby('Product Category')[['Price per Unit', 'Quantity']].corr().iloc[0::2, -1]
price_quantity_corr = price_quantity_corr.reset_index().drop('level_1', axis=1)
price_quantity_corr.columns = ['Product Category', 'Price_Quantity_Correlation']

## 6. Análisis Temporal

# Ventas por mes
monthly_sales = df.groupby(['Year', 'Month']).agg({
    'Total Amount': 'sum',
    'Quantity': 'sum',
    'Transaction ID': 'count'
}).reset_index()
monthly_sales.columns = ['Year', 'Month', 'Total_Sales', 'Total_Quantity', 'Transaction_Count']

# Crecimiento mensual
monthly_sales['Monthly_Growth'] = monthly_sales['Total_Sales'].pct_change() * 100

## 7. Resumen Estadístico

# Estadísticas descriptivas generales
summary_stats = df.describe(include='all')

# Mostrar algunas conclusiones clave
print("\n=== Resumen de Análisis ===")
print(f"1. Período de datos: {df['Date'].min().date()} a {df['Date'].max().date()}")
print(f"2. Total de ventas: ${df['Total Amount'].sum():,.2f}")
print(f"3. Transacciones totales: {df.shape[0]}")
print(f"4. Clientes únicos: {df['Customer ID'].nunique()}")
print("\nDistribución por categoría de producto:")
print(df['Product Category'].value_counts(normalize=True))
print("\nDistribución por categoría de venta:")
print(df['Sales_Category'].value_counts(normalize=True))
```

Distribución de Ventas por Categoría:

Las categorías de productos muestran diferentes patrones de ventas, con algunas generando más ingresos totales y otras con mayores volúmenes de transacciones.

Diferencias Demográficas:

```python
###Proyecto 1 – Parte V (Core)
##Análisis y Predicción de Ventas en una Tienda de Retail

###Parte V "Exploración y Visualización de Datos Avanzada / Proyecto I- Parte V (Core)"

# Asegúrate de que las librerías necesarias estén importadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Se asume que df ya está cargado y preparado desde la Parte IV
# Si estás ejecutando esta parte por separado, asegúrate de cargar y preparar df
# Ejemplo de carga y preparación básica si es necesario ejecutar de forma independiente:
# try:
#     df = pd.read_csv('retail_sales_dataset.csv')
#     df['Date'] = pd.to_datetime(df['Date'])
# except FileNotFoundError:
#     print("Error: retail_sales_dataset.csv not found. Please make sure the file is in the correct directory.")
# except Exception as e:
#     print(f"An error occurred while loading or processing the data: {e}")

# Asegurarse de que las columnas clave para el análisis existan
# Si df no fue completamente procesado en celdas anteriores
if 'WeekdayName' not in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df['WeekdayName'] = df['Date'].dt.day_name()

if 'Age_Group' not in df.columns:
    bins = [0, 25, 40, 55, 100]
    labels = ['<25', '25-40', '40-55', '55+']
    df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

if 'Customer_Type' not in df.columns:
    customer_frequency = df['Customer ID'].value_counts().reset_index()
    customer_frequency.columns = ['Customer ID', 'Purchase_Count']
    df = df.merge(customer_frequency, on='Customer ID', how='left')
    bins = [0, 1, 3, float('inf')]
    labels = ['One-Time', 'Occasional', 'Frequent']
    df['Customer_Type'] = pd.cut(df['Purchase_Count'],
                                bins=bins,
                                labels=labels)

# Recrear customer_value si es necesario para análisis de clientes
try:
    customer_value['Value_Category'].head() # Check if it exists
except NameError:
    customer_value = df.groupby('Customer ID').agg({
        'Total Amount': 'sum',
        'Quantity': 'sum',
        'Transaction ID': 'count'
    }).reset_index()
    customer_value.columns = ['Customer ID', 'Total_Spent', 'Total_Items', 'Transaction_Count']
    customer_value['Value_Category'] = pd.qcut(customer_value['Total_Spent'],
                                             q=4,
                                             labels=['Low', 'Medium', 'High', 'VIP'])


print("DataFrame listo para análisis avanzado.")
print(df.info())

## 1. Análisis de Correlación Avanzado

# Seleccionar columnas numéricas para la matriz de correlación
numeric_cols = df.select_dtypes(include=np.number).columns
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación entre Variables Numéricas', fontsize=16)
plt.show()

# Análisis de correlación entre precio y cantidad por categoría de producto
print("\nCorrelación entre Precio y Cantidad por Categoría de Producto:")
price_quantity_corr_category = df.groupby('Product Category')[['Price per Unit', 'Quantity']].corr().unstack().iloc[:, 1]
print(price_quantity_corr_category)


## 2. Visualización con Subplots y Elementos Avanzados

# Subplots: Ventas totales por mes y por día de la semana
fig, axes = plt.subplots(2, 1, figsize=(12, 10))
fig.suptitle('Análisis Temporal de Ventas Avanzado', fontsize=16)

# Gráfico 1: Ventas Mensuales
monthly_sales = df.groupby(pd.Grouper(key='Date', freq='M'))['Total Amount'].sum()
monthly_sales.plot(marker='o', ax=axes[0], linestyle='-')
axes[0].set_title('Tendencia de Ventas Mensuales', fontsize=14)
axes[0].set_xlabel('Fecha')
axes[0].set_ylabel('Ventas Totales ($)')
axes[0].grid(True, linestyle='--', alpha=0.6)

# Anotar el pico de ventas (ejemplo: mes con mayores ventas)
peak_month = monthly_sales.idxmax()
peak_sales = monthly_sales.max()
axes[0].annotate(f'Pico: ${peak_sales:,.0f}',
                 xy=(peak_month, peak_sales),
                 xytext=(peak_month + pd.DateOffset(months=2), peak_sales * 0.9),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
                 fontsize=10)


# Gráfico 2: Ventas por Día de la Semana
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_sales = df.groupby('WeekdayName')['Total Amount'].sum().reindex(weekday_order)
weekday_sales.plot(kind='bar', ax=axes[1], color=sns.color_palette("viridis", len(weekday_order)))
axes[1].set_title('Ventas por Día de la Semana', fontsize=14)
axes[1].set_xlabel('Día de la Semana')
axes[1].set_ylabel('Ventas Totales ($)')
axes[1].grid(axis='y', linestyle='--', alpha=0.6)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

# Anotar el día con mayores y menores ventas
max_day = weekday_sales.idxmax()
max_day_sales = weekday_sales.max()
min_day = weekday_sales.idxmin()
min_day_sales = weekday_sales.min()

axes[1].annotate(f'Máx: ${max_day_sales:,.0f}',
                 xy=(weekday_sales.index.get_loc(max_day), max_day_sales),
                 xytext=(weekday_sales.index.get_loc(max_day) + 0.5, max_day_sales * 0.9),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
                 fontsize=10)

axes[1].annotate(f'Mín: ${min_day_sales:,.0f}',
                 xy=(weekday_sales.index.get_loc(min_day), min_day_sales),
                 xytext=(weekday_sales.index.get_loc(min_day) + 0.5, min_day_sales * 1.1),
                 arrowprops=dict(facecolor='red', shrink=0.05, width=1, headwidth=8),
                 fontsize=10, color='red')


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajustar el layout para dejar espacio para el título principal
plt.show()


# Subplots: Distribución de ventas por Categoría de Venta y Tipo de Cliente (pie charts)
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
fig.suptitle('Distribución de Ventas y Clientes', fontsize=16)

# Gráfico 1: Distribución por Categoría de Venta
# Asegurarse de que la columna 'Sales_Category' exista
if 'Sales_Category' not in df.columns:
    percentiles = df['Total Amount'].quantile([0.33, 0.66]).values
    df['Sales_Category'] = pd.cut(df['Total Amount'],
                                 bins=[0, percentiles[0], percentiles[1], float('inf')],
                                 labels=['Low', 'Medium', 'High'])

df['Sales_Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, ax=axes[0], colors=sns.color_palette('pastel'))
axes[0].set_title('Distribución de Ventas por Categoría', fontsize=14)
axes[0].set_ylabel('') # Ocultar etiqueta y del pie chart
axes[0].legend(title="Categoría", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # Añadir leyenda

# Gráfico 2: Distribución por Tipo de Cliente
df['Customer_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, ax=axes[1], colors=sns.color_palette('cool'))
axes[1].set_title('Distribución de Clientes por Tipo', fontsize=14)
axes[1].set_ylabel('') # Ocultar etiqueta y del pie chart
axes[1].legend(title="Tipo de Cliente", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1)) # Añadir leyenda

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


# Subplots: Boxplots de Monto Total por Género y por Categoría de Producto
fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.suptitle('Distribución de Monto Total por Segmento', fontsize=16)

# Gráfico 1: Monto Total por Género
sns.boxplot(data=df, x='Gender', y='Total Amount', ax=axes[0], palette='Set2')
axes[0].set_title('Distribución de Monto Total por Género', fontsize=14)
axes[0].set_xlabel('Género')
axes[0].set_ylabel('Monto Total ($)')
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# Gráfico 2: Monto Total por Categoría de Producto
sns.boxplot(data=df, x='Product Category', y='Total Amount', ax=axes[1], palette='Set3')
axes[1].set_title('Distribución de Monto Total por Categoría de Producto', fontsize=14)
axes[1].set_xlabel('Categoría de Producto')
axes[1].set_ylabel('Monto Total ($)')
axes[1].grid(axis='y', linestyle='--', alpha=0.6)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right') # Rotar etiquetas del eje x

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


## 3. Visualización de Relaciones con Elementos Avanzados

# Scatter plot con anotación para un punto de datos específico (ejemplo: la transacción de mayor monto)
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='Price per Unit', y='Quantity', hue='Product Category', size='Total Amount', sizes=(30, 300), alpha=0.7)

plt.title('Relación entre Precio, Cantidad y Monto Total por Categoría', fontsize=16)
plt.xlabel('Precio por Unidad ($)')
plt.ylabel('Cantidad')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Categoría de Producto', bbox_to_anchor=(1.05, 1), loc='upper left') # Ajustar posición de la leyenda

# Encontrar la transacción de mayor monto
highest_sale = df.loc[df['Total Amount'].idxmax()]

# Anotar la transacción de mayor monto
plt.annotate(f'Venta Máxima:\n${highest_sale["Total Amount"]:,.0f}\n({highest_sale["Product Category"]})',
             xy=(highest_sale['Price per Unit'], highest_sale['Quantity']),
             xytext=(highest_sale['Price per Unit'] + df['Price per Unit'].std() * 0.5, highest_sale['Quantity'] + df['Quantity'].std() * 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
             fontsize=9, ha='left', va='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1))


plt.show()


print("\nAnálisis Avanzado y Visualizaciones Completadas.")

# --- Instrucciones Adicionales (No son código Python, solo recordatorios) ---
# 1. Generar un Tag en GitHub:
#    - Crea una nueva rama: `git checkout -b feature/advanced-eda-viz`
#    - Añade tus cambios: `git add .`
#    - Haz commit: `git commit -m "Completa Parte V: EDA y Visualización Avanzada"`
#    - Sube la rama: `git push origin feature/advanced-eda-viz`
#    - Ve a GitHub, crea un Pull Request desde esta rama a `main` (o tu rama principal), y fusiona.
#    - Una vez fusionado, crea un tag: `git tag v1.5`
#    - Sube el tag: `git push origin v1.5`

# 2. Realizar una Presentación:
#    - Prepara slides con los puntos mencionados en las instrucciones.
#    - Incluye los gráficos generados en esta parte V y explícalos.
#    - Resume los hallazgos clave y las conclusiones.
```
